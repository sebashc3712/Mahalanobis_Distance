```python
import pandas as pd
from scipy.spatial import distance
import numpy as np
```

```python
cat=pd.read_excel("Input Python.xlsx",sheets="Cat")
```

```python
cat.head()
```

```python
DS=pd.read_excel("Input Python DS.xlsx",sheets="DS")
```

```python
DS.head()
```

```python
columns_d=DS.columns.insert(0,"Category")
matrix=pd.DataFrame(columns=columns_d)
```

```python
matrix["Category"]=cat.columns
```

```python
matrix=matrix.fillna(0)
matrix=matrix.set_index("Category")
matrix
```

```python
for i in cat.columns:
    
    for j in DS.columns:

        df=pd.DataFrame()
        df[i]=cat[i]
        df[j]=DS[j]

        cov=np.cov(df)

        matrix.loc[i,j]=distance.mahalanobis(df[i],df[j],cov.T)
```

```python
matrix
```

```python
matrix.to_excel("Output_Mahalanobis.xlsx")
```
